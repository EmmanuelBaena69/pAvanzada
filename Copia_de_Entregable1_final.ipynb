## **Gradiente descendente 1D**

En la siguiente celda se observa la función mátematica que debe optimizar. El valor mínimo exacto es de 0.0, sin embargo, se permite llegar a una aproximación del mínimo de < 0.2.


```python
funMat_GD1D = lambda x: (1 - x)**2 + 100 * (x**2)**2
```

```python
fmini = funMat_GD1D(0)
mini = 0
print("Resultados: argumento que minimiza = {:f} y mínimo en {:f}".format(mini,fmini))
```

```python
def GradDescend(fun, x_inicial, tasa_aprendizaje, num_iters):
    """GradDescend debe ser una función de Python que reciba como parámetros
       de entrada:
        -fun: la función matemática de la que se desea saber
              el mínimo y argumento que minimiza
        -x_inicial: valor inicial de x
        -tasa_aprendizaje: tasa de aprendizaje
        -num_iters: número máximo de iteraciones

      Como parámetros de salida debe retornar el mínimo (fmini) y el argumento que
      minimiza (mini), estimados a partir del método de gradiente descendente. Además,
      la función debe imprimir el valor de x y f´(x), [gradiente], f(x) en cada iteración.
    """
    h = 1e-6
    x = x_inicial
    for i in range(num_iters):
      gradiente = (fun(x + h) - fun(x)) / h  # Aproximación de la derivada
      x = x - tasa_aprendizaje * gradiente  # Actualizar
      print (f"iteracion {i+1}: x = {x:.6f}, f(x) = {fun(x):.6f}, gradiente = {gradiente:.6f}")

    fmini = fun(x)
    mini = x

    return fmini, mini
```

```python
# Selección de párametros:
tasa_aprendizaje = 0.001        # Indicar el valor deseado de tasa de aprendizaje
num_iters = 500              # Indicar el número máximo de iteraciones
x_inicial = 1.8             # Se recomienda no modificar el punto inicial, pero puede ensayar con un valor entre [-1.9, 1.9]
```

```python
# Llamar al gradiente descendente
fmini, mini = GradDescend(funMat_GD1D, x_inicial, tasa_aprendizaje, num_iters)
print("Resultados Gradiente Descendente: argumento que minimiza = {:f} y mínimo en {:f}".format(mini,fmini))
```

## **Gradiente descendente en 2D ($f(x,y)$)**

En la siguiente celda se observa la función mátematica que debe optimizar. El valor mínimo exacto es de (0,0), sin embargo, se permite llegar a una aproximación del mínimo de < 0.2.

```python
funMat_GD2D = lambda x,y: (1 - x)**2 + 100 * (y - x**2)**2
```

```python
def GradDescend2D(fun, x_inicial, y_inicial, tasa_aprendizaje, num_iters):
    """GradDescend debe ser una función de Python que reciba como parámetros
       de entrada:
        -fun: la función matemática de la que se desea saber
              el mínimo y argumento que minimiza
        -x_inicial: valor inicial de x
        -y_inicial: valor inicial de y
        -tasa_aprendizaje: tasa de aprendizaje
        -num_iters: número máximo de iteraciones

      Como parámetros de salida debe retornar el mínimo (fmini) y los argumentos (x,y) que
      minimizan (mini_x, mini_y), estimados a partir del método de gradiente descendente. Además,
      la función debe imprimir el valor de <x,y> y f´(x) [gradiente], f(x,y) en cada iteración.
    """
    h = 1e-6
    x = x_inicial
    y = y_inicial

    for i in range(num_iters):
        gradiente_x = (fun(x + h, y) - fun(x, y)) / h  # Aproximación de la derivada en x
        gradiente_y = (fun(x, y + h) - fun(x, y)) / h  # Aproximación de la derivada en y
        x = x - tasa_aprendizaje * gradiente_x  # Actualizar
        y = y - tasa_aprendizaje * gradiente_y  # Actualizar

        f_val = fun(x, y)  # Evaluar función en el nuevo punto

        print (f"Iteración {i}: x = {x:.6f}, y = {y:.6f}, f(x, y) = {f_val:.6f}, gradiente = ({gradiente_x:.6f}, {gradiente_y:.6f})")

    fmini = fun(x, y)
    mini_x = x
    mini_y = y

    return fmini, mini_x, mini_y
```

```python
# Selección de párametros:
tasa_aprendizaje = 0.001         # Indicar el valor deseado de tasa de aprendizaje
num_iters = 1001               # Indicar el número máximo de iteraciones
x_inicial = 1.5             # Se recomienda no modificar el punto inicial, pero puede ensayar con un valor entre [-1.9, 1.9]
y_inicial = 1.5             # Se recomienda no modificar el punto inicial, pero puede ensayar con un valor entre [-1.9, 1.9]
```

```python
# Llamar al gradiente descendente2D
fmini, mini_x, mini_y = GradDescend2D(funMat_GD2D, x_inicial, y_inicial, tasa_aprendizaje, num_iters)
print("Resultados Gradiente Descendente: argumento que minimiza = ({:f},{:f}) y mínimo en {:f}".format(mini_x,mini_y,fmini))
```
